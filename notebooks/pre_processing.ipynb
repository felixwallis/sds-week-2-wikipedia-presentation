## Creating the Parent Dataset
**Loopping through and merging all data points and sections together**

```python
# Querying all the XML files for our two Politicians
putin_xml_files = wikihelper.get_xml_file_paths('Data/Vladimir Putin')
xi_xml_files = wikihelper.get_xml_file_paths('Data/Xi Jinping')
```

```python
print(putin_xml_files[:3])
print(xi_xml_files[:3])
```

```python
# Initializing an empty DataFrame
combined_df = pd.DataFrame()

count = 0

for file_i in putin_xml_files + xi_xml_files:
    # Initialize the parser
    parser = WikiXMLParser()

    # Extract sections to DataFrame
    df_article = parser.extract_sections_to_dataframe(file_i)

    # Concatenate the new DataFrame with the combined DataFrame
    combined_df = pd.concat([combined_df, df_article], ignore_index=True)

    # Increment the count
    count += 1
    
    # Print every 100 files processed
    if count % 100 == 0:
        print(f"{count} XML Files Churned ....")
```

```python
combined_df.to_feather('wikiarticles_seg_data.feather')
```

```python
combined_df.to_csv('wikiarticles_seg_data.csv')
```

```python
combined_df['category'] = None  # Initialize new column

# Iterate over the keys in the JSON mapping
for category, values in topic_groups.items():
    combined_df.loc[combined_df['title'].isin(values), 'category'] = category
```

```python
combined_df.category.isna().sum()
```

```python
combined_df[combined_df.category.isna()].title.value_counts()
```

```python
# Drop rows where the 'category' column is NaN
combined_df = combined_df.dropna(subset=['category'])
```
